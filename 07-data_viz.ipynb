- Draw a basic graph with 1 variable (count plot)

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#set a theme
sns.set_theme(style='ticks',color_codes=True)

#load dataset from seaborn
khasti = sns.load_dataset('titanic')


# #draw basic graph with 1 varibale(count plot)
sns.countplot(x='sex',data=khasti)

#plot show
plt.show()
```

- Draw a basic graph with 2 variable and add 'hue' function (count plot)

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#set a theme
sns.set_theme(style='ticks',color_codes=True)

#load dataset from seaborn
khasti = sns.load_dataset('titanic')


# #draw basic graph with 2 varibale(count plot)
sns.countplot(x='sex',hue='class',data=khasti)

#show plot
plt.show()
```

- Draw a basic graph with 2 variable and add 'hue' function & add 'title' (count plot)

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#set a theme
sns.set_theme(style='ticks',color_codes=True)

#load dataset from seaborn
khasti = sns.load_dataset('titanic')


# #draw basic graph with 2 varibale(count plot)
sns.countplot(x='sex',hue='class',data=khasti)

#add a title
plt.title('Ali ka count plot')

#show plot
plt.show()
```
