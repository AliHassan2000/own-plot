## Draw a line plot

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset of iris from seaborn
phool=sns.load_dataset('iris')


# #draw a line plot
sns.lineplot(x='sepal_length',y='sepal_width',data=phool)

#show plot
plt.show()

```

- Draw a line graph and add tittle

- Draw a line graph  & add x and y limits

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset of iris from seaborn
phool=sns.load_dataset('iris')


# #draw a line plot
sns.lineplot(x='sepal_length',y='sepal_width',data=phool)

#add title
plt.title('phoolo ka line plot')

#Adding a limits
plt.xlim(2)
plt.ylim(1)

#show plot
plt.show()
```

## Bar plots

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset 
phool=sns.load_dataset('iris')

#set style
sns.set_style('darkgrid')

#draw line plot
sns.barplot(x='species',y='sepal_width',data=phool)
plt.show()
```

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset 
khasti=sns.load_dataset('titanic')

#draw line plot
sns.barplot(x='sex',y='alone',hue='who',data=khasti)
plt.show()
```

- how to change order (order mean x axix ma kis side pa kon c chease rahkni h)

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset 
khasti=sns.load_dataset('titanic')

#draw line plot
sns.barplot(x='sex',y='alone',hue='who',data=khasti,order=['female','male']) # order=['female,male] (change the order of x axix)
plt.show()
```

- change the color 

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset 
khasti=sns.load_dataset('titanic')

#draw line plot
sns.barplot(x='sex',y='alone',hue='who',data=khasti,order=['female','male'],color='red') #color='red (this code change the color of plot)
plt.show()
```

- Remove the error bar 

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset 
khasti=sns.load_dataset('titanic')

#draw line plot
sns.barplot(x='sex',y='alone',hue='who',data=khasti,order=['female','male'],color='red',ci=None) # ci(confedantional interval) this command remove the error color
plt.show()
```

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset 
khasti=sns.load_dataset('titanic')

#draw line plot
sns.barplot(x='sex',y='alone',hue='who',data=khasti,order=['female','male'],color='red',ci=None,
palette='pastel') 
plt.show()
```

- estimator Use
  - estimator are use for mean and median   

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import mean   #import a mean function from numpy libraries

#load dataset 
khasti=sns.load_dataset('titanic')

#draw line plot
sns.barplot(x='class',y='fare',hue='sex',data=khasti,estimator=mean)
plt.show()
```

- color saturation change

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset 
khasti=sns.load_dataset('titanic')

#draw line plot
sns.barplot(x='sex',y='alone',hue='who',data=khasti,estimator=mean,saturation=1)
```
